```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('lan_img1.jpg.jpeg')
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
blur = cv2.GaussianBlur(gray, (5, 5), 0)
```

```python
edges = cv2.Canny(blur, 100, 200)
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
plt.imshow(edges, cmap='gray')
plt.title("Canny Edge Detection")
plt.axis('off')
plt.show()
```

```python
edges_low = cv2.Canny(blur, 50, 100)
edges_high = cv2.Canny(blur, 150, 300)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(edges_low, cmap='gray')
plt.title("Low Threshold")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(edges_high, cmap='gray')
plt.title("High Threshold")
plt.axis('off')

plt.show()
```
